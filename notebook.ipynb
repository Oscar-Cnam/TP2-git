```python
stock = {
"pains au chocolat": 500,
"croissants": 400,
"baguettes": 500,
"beignets": 200
}
```

```python
print(stock)
```

```python
def fabriquer(produit, quantite):
    stock[produit] = quantite
    return stock
```

```python
fabriquer("cookies", 20)
```

```python
quatite_tresorerie = 0
```

```python
def vendre(produit, quantite, prix):
    prix_vente = 0
    stock[produit] -= quantite
    prix_vente += prix*quantite
    return(prix_vente)
```

```python
quatite_tresorerie += vendre("croissants", 10, 1)
```
