```python
stock = {
"pains au chocolat": 500,
"croissants": 400,
"baguettes": 500,
"beignets": 200
}
```

```python
print(stock)
```

```python
def fabriquer(produit, quantite):
    stock[produit] = quantite
    return stock
```

```python
fabriquer("cookies", 20)
```

```python
def vendre(produit, quantite, prix):
    stock[produit] -= quantite
    return stock
```

```python
vendre("beignets", 5, 2.5)
```
